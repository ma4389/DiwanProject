```python
import requests
import json
from bs4 import BeautifulSoup
import csv
```

```python
res = requests.get('https://diwanegypt.com/product-category/books/lifestyle-books/calendars-diaries-lifestyle-books/')
```

```python
soup = BeautifulSoup(res.text , 'html.parser')
```

```python
all_pages = soup.find_all('a' , attrs = {'class' : "page-numbers"})
```

```python
all_pages
```

```python
for page in all_pages:
    print(page.get('href'))
```

```python
all_a = soup.find_all('a' , attrs = {'class' : "woocommerce-LoopProduct-link woocommerce-loop-product__link"})
```

```python
for a in all_a:
    print(a.get('href'))
```

```python
for a in all_a:
    print(a.get_text().strip())
```

```python
with open('Diwan.csv', mode='w', newline= '', encoding= 'utf-8') as f:
    fieldnames = ['BookNameprice','BookUrl']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for a in soup.find_all('a', attrs={'class': "woocommerce-LoopProduct-link woocommerce-loop-product__link"}):
        BookNameprice = a.get_text().strip()
        BookUrl= a.get('href')
        writer.writerow({'BookNameprice': BookNameprice, 'BookUrl': BookUrl})
```
